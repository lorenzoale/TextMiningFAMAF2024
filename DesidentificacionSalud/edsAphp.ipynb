```python
from spacy import displacy
# taken from eds github 
def render_results(doc):
    html = displacy.render(
        doc,
        style="ent",
        options={
            "colors": {
                "ADRESSE": "#1f77b4",
                "DATE": "#aec7e8",
                "DATE_NAISSANCE": "#ff7f0e",
                "HOPITAL": "#ffbb78",
                "IPP": "#2ca02c",
                "MAIL": "#98df8a",
                "NDA": "#d62728",
                "NOM": "#ff9896",
                "PRENOM": "#9467bd",
                "SECU": "#c5b0d5",
                "TEL": "#8c564b",
                "VILLE": "#c49c94",
            }
        },
    )
    data = []
    for ent in doc.ents:
        d = dict(
            start=ent.start_char,
            end=ent.end_char,
            text=ent.text,
            label=ent.label_,
            normalized_value=str(ent._.value or ""),
        )
    
        data.append(d)
```
